```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
import os
```

```python
path = 'D:/asad/Detailed-Deep-Learning-Program-main/flowers'
flower_type = os.listdir(path)
print(flower_type)
len(flower_type)
```

```python
image = []
label = []
for i in flower_type:
    path_dir = path + '/'+str(i)+'/'
    
    print(path_dir)
    for f in os.listdir(path_dir):
        if f.endswith('.jpg'):
            
            image.append(str(path_dir+'/'+ f))
            label.append(str(i))
```

# # build model

```python
import tensorflow as tf
import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Conv2D, Flatten, Dense
```

```python
from tensorflow.keras.layers import Conv2D, Flatten, Dense
model = keras.Sequential()
model.add(keras.layers.Flatten(input_shape=(60,60,3)))
model.add(keras.layers.Dense(256, activation='relu'))
model.add(keras.layers.Dense(2,activation='softmax'))
model.summary()
```

```python

import keras
from tensorflow.keras.layers import Conv2D, Flatten, Dense
input_shape = (60, 60, 3)
model_g = keras.Sequential(
    [
        keras.layers.Flatten(input_shape=(60,60,3)),
        keras.layers.Dense(256, activation="relu"),
        keras.layers.Dense(2, activation="softmax"),
    ]
)
model_g.summary()
```

```python
model.compile(
    optimizer=keras.optimizers.RMSprop(),  # Optimizer
    # Loss function to minimize
    loss=tf.keras.losses.BinaryCrossentropy(),
    # List of metrics to monitor
    metrics=[keras.metrics.SparseCategoricalAccuracy()],
)
```

```python
keras.losses.Sparseb
```

```python
print("Fit model on training data")
history = model.fit(x_train, y_train, epochs=2)
    
```

```python
history.history
```

```python
# Evaluate the model on the test data using `evaluate`
print("Evaluate on test data")
results = model.evaluate(x_test, y_test, batch_size=128)
print("test loss, test acc:", results)
```
