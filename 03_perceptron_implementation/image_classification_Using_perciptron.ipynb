```python
import os
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
tf.compat.v1.disable_eager_execution()
```

```python
#dataset path
path = 'D:/asad/Detailed-Deep-Learning-Program-main/flowers//'
flower_type = os.listdir(path)
flower_type
```

```python
img_size = 60
image =[]
label =[]


for i in flower_type:
    data_path = path+ str(i) # enter to folder
    print(data_path)
    filenames = [i for i in os.listdir(data_path) if i.endswith('.jpg')]
    print(filenames)
    for f in filenames:
        img = cv2.imread(data_path+'/'+f)
        img = cv2.resize(img,(img_size ,img_size))
        image.append(img)
        label.append(i)
```

```python
images = np.array(image)
images = images.astype('float32')/255
images
```
